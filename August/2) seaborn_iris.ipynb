```python
%matplotlib inline
import pandas as pd
import seaborn as sns
iris = sns.load_dataset('iris')
iris
```

```python
iris.describe()
```

```python
sns.lmplot(data=iris, x="sepal_length", y="sepal_width")
```

```python
sns.lmplot(data=iris, x="petal_length", y="petal_width")
```

petal length 클수록 width도 큼
