```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib 
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.describe()
```

total_bill, tip, sex, smoker, day, time, size
- total_bill과 tip
- sex, tip
- smoker, tip
- day, time 
- size 

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker",palette="Accent")
```

- total_bill - tip 상관관계 있음 (전체 비용 많을수록 팁 많이줌)
- smoker - 비흡연자가 흡연자보다 관련성 더 큼

```python
sns.barplot(data=tips, x="sex", y="tip")
```

```python
sns.barplot(data=tips, x="day", y="tip", hue="time")
```

- lunch보다 dinner 때 팁 많이줌 
- 다른 요일보다 일요일이 더 높음

```python
sns.countplot(data=tips, x="day")
```

- tips 데이터상으론 주말이 평일보다 팁 주는 빈도가 더 높음

```python
sns.stripplot(data = tips, x="day", y="total_bill", hue="time")
```

- 금요일의 경우- lunch보다 dinner 때 팁 비용 높음
- 목요일에 dinner 데이터가 한개밖에 없어서 비교하기가 어렵

```python
sns.countplot(data=tips, x="size")
```

```python
sns.stripplot(data=tips, x = "size" , y="tip")
```

```python
sns.catplot(data=tips, x="size" , y="tip", hue="smoker", kind="swarm")
```

```python
sns.catplot(data=tips, x="size" , y="tip", hue="sex", kind="swarm")
```

남자 & 비흡연자가 팁 주는 금액이 더 큰 것을 알 수 있다.

```python
sns.catplot(data=tips, x="size" , y="tip", hue="day", kind="swarm")
```

주말이 평일에 비해 팁의 빈도뿐아니라 금액 또한 높음
